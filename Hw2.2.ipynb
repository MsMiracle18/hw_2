```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv
df = pd.read_csv("2017_jun_final.csv")
print(df)
```

```python
data = df[['Город', 'Зарплата.в.месяц']]
avg_salary_by_city = data.groupby('Город')['Зарплата.в.месяц'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Город', y='Зарплата.в.месяц', data=avg_salary_by_city, palette='viridis')
plt.title('Середня зарплата за містами')
plt.xlabel('Місто')
plt.ylabel('Середня Зарплата.в.месяц')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
sum = df.isnull().sum()
print(sum)
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця Язык.программирования
df_tmp = df.isna().any()
col_not_del = ['Язык.программирования']
df_tmp[col_not_del] = False
df.drop(df_tmp.index[df_tmp], axis=1, inplace=True)
print(df)
```

```python
#Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
sum = df.isnull().sum()
print(sum)
```

```python
#Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
df.dropna()
print(df)
```

```python
#Визначте новий розмір таблиці за допомогою методу shape
rows, columns = df.shape
print(rows)
print(columns)
```

```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = df[df['Язык.программирования'] == 'Python'].copy()
print(python_data)
```

```python
#Визначте розмір таблиці python_data за допомогою методу shape
rows, columns = python_data.shape
print(rows)
print(columns)
```

```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
group_df = python_data.groupby("Должность")
print(group_df.groups)
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
new_df = group_df.agg({"Зарплата.в.месяц":["min","max"]})
print(new_df)
```

```python
date = ["Junior Software Engineer ","Senior Software Engineer","Software Engineer","System Architect","Technical Lead"]
fig, axs = plt.subplots(2, 1)
axs[0].plot(date, new_df["Зарплата.в.месяц"]["min"], label='Мінімальна зарплата')
axs[1].plot(date, new_df["Зарплата.в.месяц"]["max"], label='Максимальна зарплата')

axs[0].set_title('Мінімальна', fontsize=10)
axs[1].set_title('Максимальна', fontsize=10)
for ax in axs:
    ax.tick_params(axis='x', rotation=90)

plt.subplots_adjust(hspace=2.5)
fig.suptitle('Мінімальна та максимальна зарплата на певній посаді', fontsize=15)
plt.show()
```

```python
#Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(row):
    return (row["Зарплата.в.месяц"]["min"] + row["Зарплата.в.месяц"]["max"])/2

new_df["avg"] = new_df.apply(fill_avg_salary, axis=1)
print(new_df)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='avg', y='Должность', data=new_df, color='skyblue', label='Середня')
plt.title('Зарплата за посадою')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Посада')
plt.legend()
plt.show()
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика.
print(new_df.avg.describe())
```
